<h3>SCRAPER THAT CAPTURES MOBILE PHONE/TABLET DATA ON GSMARENA</h3>

<h5>Initialize libraries and capture the different companies listed</h5>

```python
import pandas as pd
import numpy as np
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import time
import re

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = "https://www.gsmarena.com/makers.php3"         #link to the companies listed on gsmarena
main = "https://www.gsmarena.com/"
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')
table = BeautifulSoup(str(soup.find_all("div", "st-text")), 'html.parser')
phone_links = []
for hyperlinks in table.find_all('a'):              #finds all the PARTIAL-links of the companies listed on gsmarena
    print(main + hyperlinks['href'])
    phone_links.append(main + hyperlinks['href'])   #appends to a list which will contain the final link of the company
```

<h5>Check if the given company has more than 1 webpage of devices and add those links for scraping

```python
brand_links = []
for url in phone_links:
    brand = urllib.request.urlopen(url, context=ctx).read()
    page_soup = BeautifulSoup(brand, 'html.parser')

    brand_links.append(url)                                  #add the phone_link
    
    nav_data = page_soup.find(class_='nav-pages')
    if not nav_data:                                        #check if there is more 1 page
        continue
    else:
        for link in nav_data.find_all('a'):       
            print(main + link['href']) 
            brand_links.append(main + link['href'])         #if yes, add this link to the final list of links
```

<h5>Regex function that parses the device data captured

```python
def splitter(title):
        phone = {}
        phone['raw'] = title
        phone['model'] = ''
        phone['announced'] = ''
        phone['display'] = ''
        phone['SoC'] = ''
        phone['primary camera'] = ''
        phone['front camera'] = ''
        phone['battery'] = ''
        phone['storage'] = ''
        phone['RAM'] = ''


        commas = re.split(',', title)
        try:
            phone['model'] = re.split('. Announced ', title)[0]
        except:
            phone['model'] = ''
        try:
            phone['announced'] = re.split('. Announced ', title)[1].split('.')[0]
        except:
            phone['announced'] = ''
        try:
            phone['display'] = (re.search(r'Features (\d+\.?\d*)', title).group(1))
        except:
            phone['display'] = ''
        try:
            for value in commas:
                if re.search('chipset', value) is not None:
                    phone['SoC'] = value
        except:
            phone['SoC'] = ''

        try:
            for value in commas:
                if re.search('primary camera', value) is not None:
                    phone['primary camera'] = value
        except:
            phone['primary camera'] = ''
        try:
            for value in commas:
                if re.search('front camera', value) is not None:
                    phone['front camera'] = value
        except:
            phone['front camera'] = ''
        try:
            for value in commas:
                if re.search('battery', value) is not None:
                    phone['battery'] = value
        except:
            phone['battery'] = ''
        try:
            for value in commas:
                if re.search('storage', value) is not None:
                    phone['storage'] = value
        except:
            phone['storage'] = ''
        try:
            for value in commas:
                if re.search('RAM', value) is not None:
                    phone['RAM'] = value
        except:
            phone['RAM'] = ''
        print(phone)
        return phone
```

<h5>Write the required data into a csv file 

```python
import csv


for url in brand_links:
    brand = urllib.request.urlopen(url, context=ctx).read()
    page_soup = BeautifulSoup(brand, 'html.parser')              
    table = page_soup.find(class_ = 'makers')                   #Locate all the devices listed on the page
    image = table.find_all('img')                               #Locate the img tags which contains the required data
    x = BeautifulSoup(str(image), 'html.parser')
    img_tags = x.find_all('img')
    titles = [tag.get('title') for tag in img_tags]             #Store the data found for each device in a list
    for title in titles:
        data = splitter(title)
        try:

            with open(r"C:\Users\Ayushman\Desktop\ee005\Smartphones\Doc_ID_test.csv", 'a+', newline='', encoding='utf-8') as file:    #Replace the filepath as required
                writer = csv.DictWriter(file, fieldnames=data.keys())

                is_empty = file.tell() == 0
                if is_empty:
                    writer.writeheader()
                writer.writerow(data)
        except UnicodeEncodeError as e:
            print("Error encountered with title:", title)
```
